```python
# Cell 2 - Import libraries
import pandas as pd
import numpy as np
from scipy import stats as sstats
from scipy.stats import ks_2samp
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import LeaveOneOut, cross_val_predict
from sklearn.metrics import roc_auc_score, confusion_matrix, roc_curve
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Reproducibility
np.random.seed(42)

# Plot style
plt.rcParams.update({
    'font.family': 'serif',
    'font.size': 11,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'figure.dpi': 150
})

print('All libraries imported successfully.')
```

```python
# Cell 4 - Load and merge data
df  = pd.read_excel(filename, sheet_name='1. Dados Brutos', header=1)
rom = pd.read_excel(filename, sheet_name='5. ADM – Dados Brutos', header=1)

# Merge on participant name
data = df.merge(rom, on='NOME', how='left', suffixes=('', '_rom'))

# Identify group column
gcol = [c for c in data.columns if 'GRUPO' in str(c).upper()][0]

print(f'Group column    : {gcol}')
print(f'Groups found    : {list(data[gcol].unique())}')
print(f'Total participants: {len(data)}')
print(f'\nOutcome columns available:')
outcome_cols = [c for c in data.columns if any(k in str(c) for k in
                ['Δ','PRÉ','PÓS','FLEX','WHOQOL','PSEQ','ADAP','SPADI','TAMPA','PCS'])]
for c in outcome_cols:
    print(f'  {c}')
```

```python
# Cell 5 - Define Composite Response Score (CRS)
# Responder = meets at least 3 of 5 criteria (anchored to SRI >= 80%)

data['C1_WHOQOL'] = data['Δ WHOQOL']       >= 8
data['C2_PSEQ']   = data['Δ PSEQ']         >= 5
data['C3_ADAP']   = data['Δ ADAP']         <= -10
data['C4_FlexE']  = data['FLEX E\nΔ']      >= 15
data['C5_FlexD']  = data['FLEX D\nΔ']      >= 15

criteria = ['C1_WHOQOL','C2_PSEQ','C3_ADAP','C4_FlexE','C5_FlexD']
data['CRS_score'] = data[criteria].sum(axis=1)
data['CRS']       = (data['CRS_score'] >= 3).astype(int)

# Overall
total_resp = data['CRS'].sum()
total_n    = len(data)
print('=' * 50)
print('CRS — COMPOSITE RESPONSE SCORE')
print('=' * 50)
print(f'Responders  (CRS=1): {total_resp}/{total_n} ({total_resp/total_n*100:.1f}%)')
print(f'Non-responders (CRS=0): {total_n-total_resp}/{total_n} ({(total_n-total_resp)/total_n*100:.1f}%)')

# Per group
print('\nBy group:')
for grp in data[gcol].unique():
    sub = data[data[gcol] == grp]['CRS']
    print(f'  {grp}: {sub.sum()}/{len(sub)} ({sub.mean()*100:.1f}%)')

# Criteria breakdown
print('\nIndividual criteria met:')
print(data[['NOME', gcol] + criteria + ['CRS_score','CRS']].to_string(index=False))
```

```python
# Cell 6 - Define feature matrix (X) and outcome vector (y)

feature_map = {
    'Age':        'IDADE',
    'Weight_kg':  'PESO (kg)',
    'TSK_pre':    'TAMPA PRÉ',
    'PCS_pre':    'PCS PRÉ',
    'PSEQ_pre':   'PSEQ PRÉ',
    'ADAP_pre':   'ADAP PRÉ',
    'WHOQOL_pre': 'WHOQOL PRÉ',
    'SPADI_pre':  'SPADI-TOT PRÉ',
    'FlexD_pre':  'FLEX D\nPRÉ',
    'FlexE_pre':  'FLEX E\nPRÉ',
}

# Keep only features present in the dataset
valid = {k: v for k, v in feature_map.items() if v in data.columns}
feat_names = list(valid.keys())
feat_cols  = list(valid.values())

X = data[feat_cols].values.astype(float)
y = data['CRS'].values

print(f'Feature matrix X: {X.shape}  (participants × features)')
print(f'Outcome vector y: {y.shape}  — {np.bincount(y)[0]} non-responders, {np.bincount(y)[1]} responders')
print(f'\nFeatures:')
for i, (k, v) in enumerate(valid.items()):
    print(f'  {i+1:2}. {k:<15} ← column "{v}"')
```

```python
# Cell 7 - Gaussian Copula functions

def fit_gaussian_copula(data_array):
    """
    Fit a Gaussian Copula to a real data matrix.
    Returns the correlation matrix R of rank-transformed data.
    """
    n, p = data_array.shape
    # Step 1: rank-transform to uniform (empirical CDF)
    U = np.array(
        [sstats.rankdata(data_array[:, j]) / (n + 1) for j in range(p)]
    ).T
    # Step 2: probit transform (uniform -> normal)
    Z = sstats.norm.ppf(np.clip(U, 1e-6, 1 - 1e-6))
    # Step 3: estimate Pearson correlation matrix
    R = np.corrcoef(Z.T)
    np.fill_diagonal(R, 1.0)
    R = np.clip(R, -0.999, 0.999)
    return R


def sample_gaussian_copula(R, original_data, n_samples):
    """
    Generate synthetic samples from a fitted Gaussian Copula.
    Marginals are recovered via empirical quantile interpolation.
    """
    p = R.shape[0]
    # Draw from multivariate normal with correlation R
    Z_syn = np.random.multivariate_normal(np.zeros(p), R, size=n_samples)
    # Transform back to uniform
    U_syn = sstats.norm.cdf(Z_syn)
    # Recover original marginal distributions
    X_syn = np.zeros((n_samples, p))
    for j in range(p):
        sorted_vals  = np.sort(original_data[:, j])
        quantile_grid = np.linspace(0, 1, len(sorted_vals))
        X_syn[:, j]  = np.interp(U_syn[:, j], quantile_grid, sorted_vals)
    return X_syn


print('Gaussian Copula functions defined successfully.')
```

```python
# Cell 8 - Fit copula per group and generate synthetic data

# Split real data by group
g1_mask = data[gcol] == 'EXPERIMENTAL'
g0_mask = data[gcol] == 'CONTROLE'

X_g1, y_g1 = X[g1_mask], y[g1_mask]
X_g0, y_g0 = X[g0_mask], y[g0_mask]

# Fit copula separately for each group
R_g1 = fit_gaussian_copula(X_g1)
R_g0 = fit_gaussian_copula(X_g0)

# Generate 200 synthetic observations per group
N_SYN = 200
X_syn_g1 = sample_gaussian_copula(R_g1, X_g1, N_SYN)
X_syn_g0 = sample_gaussian_copula(R_g0, X_g0, N_SYN)

# Stack synthetic data from both groups
X_syn_all = np.vstack([X_syn_g1, X_syn_g0])

print(f'Real data       : {X.shape[0]} participants')
print(f'Synthetic G1    : {X_syn_g1.shape[0]} observations')
print(f'Synthetic G0    : {X_syn_g0.shape[0]} observations')
print(f'Total synthetic : {X_syn_all.shape[0]} observations')
print(f'Augmented total : {X.shape[0] + X_syn_all.shape[0]} observations')
```

```python
# Cell 9 - Assign CRS labels to synthetic data
# Uses logistic regression trained on real data to predict
# CRS probability for each synthetic observation

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit logistic model on real data
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
lr.fit(X_scaled, y)

# Predict CRS probability for synthetic observations
X_syn_scaled = scaler.transform(X_syn_all)
prob_syn     = lr.predict_proba(X_syn_scaled)[:, 1]

# Assign labels via stochastic sampling from predicted probabilities
np.random.seed(42)
y_syn = np.array([np.random.binomial(1, p) for p in prob_syn])

# Combine real + synthetic
X_aug = np.vstack([X, X_syn_all])
y_aug = np.concatenate([y, y_syn])

print('CRS label assignment for synthetic data:')
print(f'  Synthetic responders : {y_syn.sum()}/{len(y_syn)} ({y_syn.mean()*100:.1f}%)')
print(f'  Real responders      : {y.sum()}/{len(y)} ({y.mean()*100:.1f}%)')
print(f'\nAugmented dataset:')
print(f'  X_aug shape : {X_aug.shape}')
print(f'  y_aug dist  : {np.bincount(y_aug)[0]} non-responders, {np.bincount(y_aug)[1]} responders')
```

```python
# Cell 10 - Kolmogorov-Smirnov fidelity test
# Confirms synthetic distributions are statistically
# indistinguishable from real data (p > 0.05 = pass)

print('Kolmogorov-Smirnov Fidelity Test — Real vs Synthetic')
print('p > 0.05 = distributions are statistically identical')
print('-' * 52)

ks_results = []
for j, fname in enumerate(feat_names):
    stat, pval = ks_2samp(X[:, j], X_syn_all[:, j])
    status = 'PASS ✓' if pval > 0.05 else 'FAIL ✗'
    ks_results.append({
        'Feature': fname,
        'KS_stat': round(stat, 4),
        'p_value': round(pval, 4),
        'Result': status
    })
    print(f'  {status}  {fname:<15}  D={stat:.4f}  p={pval:.4f}')

ks_df = pd.DataFrame(ks_results)
passed = (ks_df['p_value'] > 0.05).sum()
print(f'\nSummary: {passed}/{len(feat_names)} features passed fidelity test')
```

```python
# Cell 11 - Random Forest + Leave-One-Out Cross-Validation
# Trained on augmented data, evaluated on real data only

rf = RandomForestClassifier(
    n_estimators=500,
    max_depth=3,
    min_samples_leaf=2,
    class_weight='balanced',
    random_state=42
)

loo = LeaveOneOut()
y_pred_proba = np.zeros(len(X))

for train_idx, test_idx in loo.split(X):
    # Real training split
    X_train_real = X[train_idx]
    y_train_real = y[train_idx]
    # Augment: real training fold + all synthetic
    X_train_aug = np.vstack([X_train_real, X_syn_all])
    y_train_aug = np.concatenate([y_train_real, y_syn])
    # Fit and predict on held-out real observation
    rf.fit(X_train_aug, y_train_aug)
    y_pred_proba[test_idx] = rf.predict_proba(X[test_idx])[:, 1]

# Binary predictions
y_pred = (y_pred_proba >= 0.5).astype(int)

# Metrics
auc        = roc_auc_score(y, y_pred_proba)
cm         = confusion_matrix(y, y_pred)
tn, fp, fn, tp = cm.ravel()
sens       = tp / (tp + fn)
spec       = tn / (tn + fp)
ppv        = tp / (tp + fp) if (tp + fp) > 0 else 0
npv        = tn / (tn + fn) if (tn + fn) > 0 else 0

print('=' * 50)
print('RANDOM FOREST — LOOCV PERFORMANCE')
print('=' * 50)
print(f'AUC          : {auc:.3f}')
print(f'Sensitivity  : {sens:.3f}  ({tp}/{tp+fn} responders correct)')
print(f'Specificity  : {spec:.3f}  ({tn}/{tn+fp} non-responders correct)')
print(f'PPV          : {ppv:.3f}')
print(f'NPV          : {npv:.3f}')
print(f'\nConfusion matrix:')
print(f'             Predicted 0  Predicted 1')
print(f'  True 0  :      {tn}            {fp}')
print(f'  True 1  :      {fn}           {tp}')
```

```python
# Cell 12 - Permutation feature importance
# Fit final model on full augmented data
rf_final = RandomForestClassifier(
    n_estimators=500,
    max_depth=3,
    min_samples_leaf=2,
    class_weight='balanced',
    random_state=42
)
rf_final.fit(X_aug, y_aug)

# Evaluate importance on REAL data only
perm = permutation_importance(
    rf_final, X, y,
    n_repeats=200,
    random_state=42,
    scoring='roc_auc'
)

# Sort by importance
sorted_idx = np.argsort(perm.importances_mean)[::-1]

print('PERMUTATION FEATURE IMPORTANCE')
print('(mean AUC decrease when feature is randomly shuffled)')
print('-' * 55)
for i in sorted_idx:
    bar = '█' * max(0, int(perm.importances_mean[i] * 500))
    print(f'  {feat_names[i]:<15} {perm.importances_mean[i]:+.4f} ± {perm.importances_std[i]:.4f}  {bar}')
```
